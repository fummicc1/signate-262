```python
import cv2
import pathlib
import subprocess
import random
import collections
```

```python
base_train_dir = pathlib.Path("/home/fummicc1/codes/signate/kfolds") / "modern" / "train" / "1" / "images"
base_val_dir = pathlib.Path("/home/fummicc1/codes/signate/kfolds") / "modern" / "val" / "1" / "images"

for base_train in list(base_train_dir.iterdir()):
    if random.randint(0, 1):
        print("target:", base_train)
        base_val = list(base_val_dir.iterdir())
        random.shuffle(base_val)
        base_val: pathlib.Path = base_val[0]
        img = base_train_dir.parent / "images" / (base_train.stem + ".jpg")
        ann = base_train_dir.parent / "annotations" / (base_train.stem + ".json")
        label = base_train_dir.parent / "labels" / (base_train.stem + ".txt")
        
        val_img = base_val_dir.parent / "images" / (base_val.stem + ".jpg")
        val_ann = base_val_dir.parent / "annotations" / (base_val.stem + ".json")
        val_label = base_val_dir.parent / "labels" / (base_val.stem + ".txt")
        subprocess.run(f"mv {img} {val_img.with_stem(img.stem)}", shell=True)
        subprocess.run(f"mv {ann} {val_ann.with_stem(ann.stem)}", shell=True)
        subprocess.run(f"mv {label} {val_label.with_stem(label.stem)}", shell=True)
        subprocess.run(f"mv {val_img} {img.with_stem(val_img.stem)}", shell=True)
        subprocess.run(f"mv {val_ann} {ann.with_stem(val_ann.stem)}", shell=True)
        subprocess.run(f"mv {val_label} {label.with_stem(val_label.stem)}", shell=True)
```
